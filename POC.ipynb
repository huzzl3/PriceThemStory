```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.support.relative_locator import locate_with


# Replace 'path/to/chromedriver' with the actual path to your chromedriver executable


# Create a new instance of the Chrome driver
driver = webdriver.Chrome()
```

```python

# Navigate to the website you want to scrape
url = 'https://www.coles.com.au/browse'
driver.get(url)
```

```python
driver.find_element(By.LINK_TEXT,"Meat & Seafood").click()
```

```python
productList = driver.find_elements(By.XPATH,"//section[@data-testid='product-tile']")
priceList = driver.find_elements(By.XPATH,"//span[@class='price__value']")
pricePerQuantityList = driver.find_elements(By.XPATH,"//div[@class='price__calculation_method']")
```

```python
product = productList[0]

productName = product.is_displayed()


print(productName)
```

```python
for product in productList:
    productName = product.find_element(By.TAG_NAME,"h2")
    print(productName.text)
```

```python
for price in priceList:
    print(price.text)
```

```python
for pricePerQuantity in pricePerQuantityList:
    print(pricePerQuantity.text)
```

```python
driver.find_element(By.XPATH,"//button[@aria-label='Go to next page']").click()
```

```python
# Remember to close the browser window when done
driver.quit()
```
